```python
class Person:
    def __init__(self, name, age, email) -> None:
        self.name = name
        self.age = age
        self.email = email
    
    @property
    def age(self) ->float:
        return self._age
    
    @age.setter
    def age (self, value:float) -> None:
        print("work")
        if not isinstance(value, (int,float)):
            raise TypeError(f"age must be number")
        if not (0 <= value  < 125):
            raise ValueError("needs to be a proper age")
        self._age = value

    @property
    def name(self)-> str:
        return self._name
    
    @name.setter
    def name (self, valuestr: str) -> None:
        print("homi")
        if not isinstance(valuestr,(str)):
            raise TypeError("fix a leter")
        self._name = valuestr
    
    @property
    def email(self) -> str:
        return self._email
    
    @email.setter
    def email (self, value4:str) -> None:
        if "@" not in value4:
            raise TypeError("you need @")
        self._email = value4
    
    def __repr__(self) -> str:
        return f"junk"

    def say_hello(self):
        print(f"hi, my name is {self.name}, im {self.age}y old, my email is {self.email}")

game = Person("game", 50, "lol@fun.com")
gam4 = Person("funz", 34, "hum@gm.com")
print(game)
print(game.age)
game.say_hello()
gam4.say_hello()
```

```python
class Student(Person):
    def say_hello(self):
        print(f"Yo, I am a student, my name is {self.name}, I am {self.age} years old, my email address is {self.email}")
        

    def study(self):
        print("study...study...study...more study")

class Teacher(Person):
    def say_hello(self):
        print(f"Yo, I am a Teacher, my name is {self.name}, I am {self.age} years old, my email address is {self.email}")
        

    def teach(self):
        print("teach...teach...teach...more teaching")


game2 = Student("lol",34,"gmail@gmail.com")
game3 = Teacher("fun", 43, "game@teach.com")
game2.study()
game2.say_hello()
game3.teach()
game3.say_hello()
```

```python
class Video:
    def __init__(self, title, genre, rating) -> None:
        self.title = title
        self.genre = genre
        self.rating = rating
    
    @property
    def title(self):
        return self._title

    @title.setter
    def title(self, value:str) -> None:
        if not isinstance(value,str):
            raise TypeError("use leters")
        self._title = value
    @property
    def genre(self):
        return self._genre

    @genre.setter
    def genre(self, value:str) -> None:
        if not isinstance(value,str):
            raise TypeError("use leters")
        self._genre = value

    @property
    def rating(self):
        return self._rating

    @rating.setter    
    def rating(self, value:float) -> None:
        if not isinstance(value,(int,float)):
            raise TypeError("use numbers")
        if not (0 <= value < 5):
            raise ValueError("the score needs to be betwen 0 and 5")
        self._rating = value
    
    def info(self)-> str:
        print(f"Video with title {self.title}, genre {self.genre} and rating {self.rating}")


class Tv_serie(Video):
    def __init__(self, title, genre, rating, num_episodes) -> None:
        super().__init__(title, genre, rating)
        self.num_episodes = num_episodes
    
    
    def info (self)-> str:
        print(f"TV series with title {self.title}, genre {self.genre}, rating {self.rating} and episodes {self.num_episodes}" )


class Movie(Video):
    def __init__(self, title, genre, rating, duration) -> None:
        super().__init__(title, genre, rating)
        self.duration = duration

    def info (self)->str:
        print(f"Movie with title {self.title}, genre {self.genre}, rating {self.rating}, duration {self.duration} minutes")


class Documentary (Video):
    def __init__(self, title, genre, rating)-> None:
        super().__init__(title, genre, rating)


pokemon = Tv_serie("Pokemon", "Cartoon", 4.5, 550)
titanic = Movie("Titanic", "Romance", 4.7, 194)
code = Documentary("The Code", "Math", 4)



for Video in tuple((pokemon, titanic, code)):
    print(Video.info())
```
